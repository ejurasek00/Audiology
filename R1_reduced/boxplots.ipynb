```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np


data_1 = (0.98, 1.01, 0.97, 1.00, 1.02, 0.99, 0.97, 1.02, 1.03, 1.00)

data = [data_1]

fig = plt.figure(figsize =(10, 7))

# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

# Creating plot
bp = ax.boxplot(data)

# Creating x axe label
ax.set_xticklabels(['maxlen=1'])

# show plot
plt.show()
```
